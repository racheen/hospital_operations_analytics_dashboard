# Data Preparation

## Standardize Categorical Features (Lowercase Conversion)

### Subtask:
Convert all categorical columns to lowercase

Reasoning: To ensure consistency across the dataset

```python
# Identify categorical columns (object dtype) excluding those we've already handled or dropped
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Convert all identified categorical columns to lowercase
for col in categorical_cols:
    df[col] = df[col].str.lower()

print("DataFrame with all categorical columns converted to lowercase:")
display(df.head())
```

```python
df.describe(include='object')
```

## Handle missing values

### Subtask:
Check for and address any missing values in the dataset.


**Reasoning**:
Check for missing values in the DataFrame.



```python
print("Missing values per column:")
print(df.isnull().sum())
```

## Convert data types

### Subtask:
Convert columns with incorrect data types, such as converting date columns to datetime objects.


**Reasoning**:
Identify and convert date columns to datetime objects.



```python
date_columns = ['Date of Admission', 'Discharge Date']
for col in date_columns:
  df[col] = pd.to_datetime(df[col])

print("Data types after conversion:")
display(df.info())
```

## Address inconsistencies

### Subtask:
Address inconsistencies identified during data exploration, like standardizing categorical entries or addressing potential misspellings if necessary.


**Reasoning**:
Examine the unique values and their counts for the specified categorical columns to identify inconsistencies.



```python
categorical_cols = ['Gender', 'Blood Type', 'Medical Condition', 'Insurance Provider', 'Admission Type', 'Test Results']

for col in categorical_cols:
  print(f"\nUnique values and counts for {col}:")
  display(df[col].value_counts())
```

## Addressing High Cardinality

### Subtask: Drop 'Name' Column

**Reasoning**:
Remove the 'Name' column as it is not needed for analysis.

```python
df = df.drop('Name', axis=1)

print("DataFrame after dropping 'Name' column:")
display(df.head())
```

### Subtask: Discuss Strategies for High Cardinality Columns ('Doctor' and 'Hospital')

**Reasoning**:
Provide potential strategies for handling the high number of unique values in the 'Doctor' and 'Hospital' columns.

```python
df['doctor_freq'] = df['Doctor'].map(df['Doctor'].value_counts())
df['hospital_freq'] = df['Hospital'].map(df['Hospital'].value_counts())

print("DataFrame with Doctor and Hospital frequencies:")
display(df[['Doctor', 'doctor_freq', 'Hospital', 'hospital_freq']].head())
```

## Feature Engineering

### Subtask: Extract Year and Day of Week from Dates

**Reasoning**:
Extract the year and day of the week from the 'Date of Admission' and 'Discharge Date' columns to analyze temporal patterns.

```python
df['Admission_Year'] = df['Date of Admission'].dt.year
df['Discharge_Year'] = df['Discharge Date'].dt.year
df['Admission_DayOfWeek'] = df['Date of Admission'].dt.day_name()
df['Discharge_DayOfWeek'] = df['Discharge Date'].dt.day_name()
df['Admission_Weekday'] = df['Date of Admission'].dt.weekday < 5 # Monday=0, Sunday=6
df['Discharge_Weekday'] = df['Discharge Date'].dt.weekday < 5

print("DataFrame with new date features:")
display(df[['Date of Admission', 'Admission_Year', 'Admission_DayOfWeek', 'Admission_Weekday', 'Discharge Date', 'Discharge_Year', 'Discharge_DayOfWeek', 'Discharge_Weekday']].head())
```

### Subtask: Calculate Admission Duration

**Reasoning**:
Calculate the duration of each patient's admission by subtracting the 'Date of Admission' from the 'Discharge Date'.

```python
df['Admission_Duration'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

print("DataFrame with Admission Duration:")
display(df[['Date of Admission', 'Discharge Date', 'Admission_Duration']].head())
display(df['Admission_Duration'].describe())
```

### Subtask: Create Age Bins

**Reasoning**:
Bin the 'Age' into 5 categories to simplify age-based analysis.

```python
df['Age_Bin'] = pd.cut(df['Age'], bins=5, labels=['13-27', '28-42', '43-57', '58-73', '74-89'])

print("DataFrame with Age Bins:")
display(df[['Age', 'Age_Bin']].head())
display(df['Age_Bin'].value_counts())
```

### Subtask: Create Age Bins (Teen, Young Adult, Adult, Senior)

**Reasoning**:
Re-display the results of the age binning with 'Teen', 'Young Adult', 'Adult', and 'Senior' categories.

```python
# Define age bins and labels based on the provided categories and age range (13-89)
# Bin 1: Teen (13-19)
# Bin 2: Young Adult (20-39)
# Bin 3: Adult (40-59)
# Bin 4: Senior (60-79)
# Bin 5: Elderly Senior (80-89) - grouping the upper age range

bins = [12, 19, 39, 59, 79, 90] # Adjusted bins to include 13 and exclude 90 as upper limit
labels = ['Teen', 'Young Adult', 'Adult', 'Senior', 'Elderly Senior'] # Adjusted labels to reflect the bins

df['Age_Bin'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False) # Use right=False to include the left edge

print("DataFrame with updated Age Bins (Teen, Young Adult, Adult, Senior, Elderly Senior):")
display(df[['Age', 'Age_Bin']].head())
display(df['Age_Bin'].value_counts())
```

### Subtask: Categorize Billing Amount

**Reasoning**:
Categorize 'Billing Amount' into 'Low', 'Medium', and 'High' based on quantiles to analyze billing patterns.

```python
df['Billing_Range'] = pd.qcut(df['Billing Amount'], q=3, labels=['Low', 'Medium', 'High'])

print("DataFrame with Billing Range:")
display(df[['Billing Amount', 'Billing_Range']].head())
display(df['Billing_Range'].value_counts())
```

### Subtask: Extract Floor from Room Number

**Reasoning**:
Extract the floor number from the 'Room Number' assuming room numbers indicate the floor (e.g., room 301 is on the 3rd floor).

```python
df['Floor'] = df['Room Number'].astype(str).str[0]

print("DataFrame with Floor:")
display(df[['Room Number', 'Floor']].head())
display(df['Floor'].value_counts())
```

### Subtask: Bin Doctor and Hospital Frequencies

**Reasoning**:
Bin the doctor and hospital frequencies to categorize them based on their occurrence in the dataset.

```python
# Bin doctor frequencies into categories (e.g., Low, Medium, High frequency)
# We can use quantiles to create bins with roughly equal numbers of observations
df['doctor_freq_bin'] = pd.qcut(df['doctor_freq'], q=4, duplicates='drop')

# Bin hospital frequencies into categories
df['hospital_freq_bin'] = pd.qcut(df['hospital_freq'], q=4, duplicates='drop')

print("DataFrame with Doctor and Hospital Frequency Bins:")
display(df[['Doctor', 'doctor_freq', 'doctor_freq_bin', 'Hospital', 'hospital_freq', 'hospital_freq_bin']].head())

print("\nValue counts for Doctor Frequency Bins:")
display(df['doctor_freq_bin'].value_counts())

print("\nValue counts for Hospital Frequency Bins:")
display(df['hospital_freq_bin'].value_counts())
```

```python
print("Min and Max Doctor Frequency:", df['doctor_freq'].min(), df['doctor_freq'].max())
print("Min and Max Hospital Frequency:", df['hospital_freq'].min(), df['hospital_freq'].max())
```

### Subtask: One-Hot Encode Hospital Frequency Bins

**Reasoning**:
One-hot encode the 'hospital_freq_bin' column to convert the categorical frequency ranges into numerical features.

```python
df = pd.get_dummies(df, columns=['hospital_freq_bin'], prefix='hospital_freq')
df = pd.get_dummies(df, columns=['doctor_freq_bin'], prefix='doctor_freq')

print("DataFrame after One-Hot Encoding Hospital Frequency Bins:")
display(df.head())
display(df.columns)
```

### Subtask: Create Feature for Doctors in Multiple Hospitals

**Reasoning**:
Create a binary feature indicating whether a doctor appeared in more than one hospital.

```python
# Group by Doctor and count the number of unique hospitals
doctor_hospital_counts = df.groupby('Doctor')['Hospital'].nunique()

# Filter for doctors who appeared in more than one hospital
doctors_in_multiple_hospitals = doctor_hospital_counts[doctor_hospital_counts > 1]

print("Doctors who appeared in more than one hospital:")
display(doctors_in_multiple_hospitals)
```

```python
# Create a list of doctors who appeared in more than one hospital
doctors_in_multiple = doctors_in_multiple_hospitals.index.tolist()

# Create a new feature in the main DataFrame
df['doctor_in_multiple_hospitals'] = df['Doctor'].apply(lambda x: x in doctors_in_multiple)

print("DataFrame with 'doctor_in_multiple_hospitals' feature:")
display(df[['Doctor', 'doctor_in_multiple_hospitals']].head())
display(df['doctor_in_multiple_hospitals'].value_counts())
```

Save the processed DataFrame to a CSV file for future use or analysis.
